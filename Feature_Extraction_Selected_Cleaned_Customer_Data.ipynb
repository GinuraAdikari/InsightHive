##Feature Engineering
###Balance
###Balance Frequency
###~Purchase + Cash Advance = Total Purchase (monetary Value)
###Installments Purchases
###~(Purchase Frequency + ONEOFF Frequency)/2 = Avg Purchase Frequency
###Cash advance frequency
###~Cash Advance Trx + Purchase Trx = Total Trx(frquency)
###Credit Limit
###Payments
###Minimum Payments
###PRC Full Payment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
SCCD = pd.read_csv('Selected_Cleaned_Customer_Data.csv')
```

```python
SCCD.head()
```

```python
SCCD.shape
```

```python
#implement codes for feature engineering process

#Create the 'Total Purchase' feature
SCCD['Total_Purchase'] = SCCD['PURCHASES'] + SCCD['CASH_ADVANCE']

#Remove the old 'PURCHASES' and 'CASH_ADVANCE' columns
SCCD.drop(columns=['PURCHASES', 'CASH_ADVANCE'], inplace=True)

#Create the 'Avg Purchase Frequency' feature
SCCD['Avg_Purchase_Frequency'] = (SCCD['PURCHASES_FREQUENCY'] + SCCD['ONEOFF_PURCHASES_FREQUENCY']) / 2

#Remove the old frequency columns
SCCD.drop(columns=['PURCHASES_FREQUENCY', 'ONEOFF_PURCHASES_FREQUENCY'], inplace=True)

#Create the 'Total Trx' feature
SCCD['Total_Trx'] = SCCD['CASH_ADVANCE_TRX'] + SCCD['PURCHASES_TRX']

#Remove the old transaction columns
SCCD.drop(columns=['CASH_ADVANCE_TRX', 'PURCHASES_TRX'], inplace=True)

# Display the first few rows of the updated dataset
SCCD.head()
```
