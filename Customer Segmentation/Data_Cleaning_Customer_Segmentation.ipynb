```python
# Fill the single missing value in CREDIT_LIMIT with the median
Customer_Data['CREDIT_LIMIT'].fillna(Customer_Data['CREDIT_LIMIT'].median(), inplace=True)
```

```python
Customer_Data.info()
```

checked Minimun Payment feature

```python
from sklearn.impute import KNNImputer

# Initialize KNN Imputer
imputer = KNNImputer(n_neighbors=5)

# Select only numerical columns for imputation
numerical_columns = Customer_Data.select_dtypes(include=['float64', 'int64']).columns

# Apply KNN Imputer to the dataset
Customer_Data[numerical_columns] = imputer.fit_transform(Customer_Data[numerical_columns])
```
