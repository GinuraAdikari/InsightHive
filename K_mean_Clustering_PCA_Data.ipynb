```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

pca_data = pd.read_csv('PCA_Data.csv')

# Apply k-means clustering with K=8
kmeans = KMeans(n_clusters=8, random_state=42)
pca_data['Cluster'] = kmeans.fit_predict(pca_data)

# Centroids of the clusters
centroids = kmeans.cluster_centers_

# Visualizing clusters (using PC1 and PC2 for simplicity)
plt.figure(figsize=(10, 6))
for cluster in range(8):
    cluster_data = pca_data[pca_data['Cluster'] == cluster]
    plt.scatter(cluster_data['PC1'], cluster_data['PC2'], label=f'Cluster {cluster}')

# Plot the centroids
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='black', marker='X', label='Centroids')

plt.title("Clusters for K=8 (Based on PC1 and PC2)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.grid()
plt.show()


```
